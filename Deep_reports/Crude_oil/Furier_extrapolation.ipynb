# FFT - Fourier Transforms

# Fourier_extrapolation with prediction plot

```python
def fourierExtrapolation(x, n_harmonics, n_predict):
    n = x.size
    t = np.arange(0, n)
    p = np.polyfit(t, x, 1)  # find linear trend in x
    x_notrend = x - p[0] * t  # detrended x
    x_freqdom = npfft.fft(x_notrend)  # detrended x in frequency domain
    f = npfft.fftfreq(n)  # frequencies
    indexes = range(n)
    # sort indexes by frequency, lower -> higher
    sorted(indexes, key=lambda i: np.absolute(f[i]))
    t = np.arange(0, n + n_predict)
    restored_sig = np.zeros(t.size)
    for i in indexes[:1 + n_harmonics * 2]:
        ampli = np.absolute(x_freqdom[i]) / n  # amplitude
        phase = np.angle(x_freqdom[i])  # phase
        restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase)
    return restored_sig + p[0] * t
```

```python
def fourier_extrapolation_start(xpd):
    n_harmonics_value = 10
    length = len(xpd)
    training_length  = int(length * 0.5)
    prediction_length  = int(length  * 0.3)                 
    columns = df.columns
    for column in columns:
        if column != df.columns[0]:
            x = xpd[str(column)]
            extrapolation = fourierExtrapolation(x, n_harmonics=n_harmonics_value, n_predict=prediction_length)
            extrapolation10 = fourierExtrapolation(x, n_harmonics=n_harmonics_value*10, n_predict=prediction_length)
            extrapolation100 = fourierExtrapolation(x, n_harmonics=n_harmonics_value*100, n_predict=prediction_length)
            extrapolation1000 = fourierExtrapolation(x, n_harmonics=n_harmonics_value*1000, n_predict=prediction_length)
                
            pl.figure(figsize=(25, 8))
            plt.title(f'fourier_extrapolation, column={column}', size=16)
            pl.plot(np.arange(0, extrapolation1000.size), extrapolation1000, 'k', label='extrapolation1000')
            pl.plot(np.arange(0, extrapolation100.size), extrapolation100, 'y', label='extrapolation100') 
            pl.plot(np.arange(0, extrapolation10.size), extrapolation10, 'g', label='extrapolation10')
            pl.plot(np.arange(0, extrapolation.size), extrapolation, 'r', label='extrapolation')
            pl.plot(np.arange(0, x.size), x, 'b', label='x', linewidth=1)
            plt.legend()
           # plt.show()
```

```python
fourier_extrapolation_start(pd.DataFrame(df[::100]))
```
