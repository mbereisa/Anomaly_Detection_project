# Sensors data visual report (no forecasting)

### Importing the data drom the file 'data_zero.csv'

```python
import matplotlib.pyplot as plt
import time
import numpy as np
import pandas as pd
%matplotlib inline
path = 'data_zero.csv'
df=pd.read_csv(path)
columns = df.columns
df.index = pd.to_datetime(df[columns[0]]/1000, unit='s')
# df.drop(columns[0], axis = 1)
delite_column = columns[0]
df = df.drop(delite_column, axis = 1) 
df
```

```python
def taking_of_nan_values_DF(df):
    # interpolation
    df = df.interpolate(method='linear', limit_direction='forward', axis=0)
    # taking of nullmi
    df = df.fillna(method='ffill')
    df = df.fillna(method='bfill')
    return df
```

```python
taking_of_nan_values_DF(df)
# df.dropna()
df.sort_index(inplace= True)
df.head()
```

### Making a simple plot of data

```python
def simple_plot(df, column, Matavimas):        
    plt.figure(figsize=(20,6))
    plt.plot(df.index, df[column])
    plt.title(Matavimas + " " + column + " real data " ,size=16)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.grid()  
```

```python
for column in columns:
    simple_plot(df, column, "Time")
```

```python
df_seconds = df.resample(rule='s').mean()
Matavimas = "seconds"
for column in columns:
    simple_plot(df_seconds, column, " seconds")
```

```python
df_min = df.resample(rule='min').mean()
```

```python
for column in columns:
    simple_plot(df_min, column, "minutes")
```

# Density plot

```python
def Density(df):
    df = df.fillna(value=df.mean()) 
    df.plot.density(figsize=(20,10)) 
```
