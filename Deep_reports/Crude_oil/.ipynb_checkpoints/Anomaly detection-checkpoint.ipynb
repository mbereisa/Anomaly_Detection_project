```python
# Laiko intervalas minutemis

laikotarpis = 10
```

```python
my_dict = {
             "std" : [],  
             "laikas" : [],
             "time_start" : [],
             "time_end" : [] 
        }
m=6000
laikas=m*laikotarpis
para=1440
laik=int(para/laikotarpis)

df0 = pd.DataFrame() 
data=df['P10215']
kiekis=laikas

for i in range(0,laik):
    if i ==0:
        b=data[:laikas]
        s=b.std()
        prad_kiekis=laikas
        s1=0
        s2=prad_kiekis
        t=df.loc[s1].time
        my_dict["std"].append(s)
        my_dict["time_start"].append(s1)
        my_dict["time_end"].append(s2)
        my_dict["laikas"].append(t)
    else:
        kiekis=kiekis+laikas
        s2=kiekis
        s1=prad_kiekis
        b=data[prad_kiekis:kiekis]
        s=b.std()
        t=df.loc[s1].time
        my_dict["std"].append(s)
        my_dict["time_start"].append(s1)
        my_dict["time_end"].append(s2)
        my_dict["laikas"].append(t)
        kiekis=kiekis
        prad_kiekis=prad_kiekis+laikas
df_final = pd.DataFrame(my_dict)

print('Data ready')
print('Creating Anomaly plot....')
ax=sns.relplot(x='laikas', y="std",kind="line",data=df_final, color='tomato',)
ax.fig.set_size_inches(14,5)
plt.xticks(rotation='vertical')
plt.xticks(rotation=80)
plt.grid()
```
